```python
import os

# 设置 JPEGImages 和 ImageSets 文件夹路径
jpeg_images_dir = './JPEGImages'
imagesets_dir = './ImageSets/main'

# 获取 JPEGImages 文件夹中的所有文件名
jpeg_images_files = os.listdir(jpeg_images_dir)

# 创建 ImageSets 文件夹（如果不存在）
os.makedirs(imagesets_dir, exist_ok=True)

# 生成 ImageSets 中的 txt 文件，例如 train.txt
train_file_path = os.path.join(imagesets_dir, 'train.txt')  # 生成 train.txt 文件
with open(train_file_path, 'w') as f:
    for image_file in jpeg_images_files:
        # 假设文件名以 .jpg 结尾，根据实际情况可能需要调整
        if image_file.endswith('.jpg'):
            image_name = os.path.splitext(image_file)[0]  # 去掉扩展名 .jpg
            f.write(image_name + '\n')

print(f"Generated {train_file_path} with {len(jpeg_images_files)} image filenames.")
```

```python
import os

def remove_spaces_from_filenames(directory, extension):
    """
    Remove spaces from filenames in the given directory with the specified extension.
    
    :param directory: The path to the directory containing the files
    :param extension: The file extension to filter by (e.g., '.jpg', '.xml')
    """
    # List all files in the directory
    files = os.listdir(directory)
    
    for filename in files:
        # Check if the file has the specified extension
        if filename.endswith(extension):
            # Remove spaces from the filename
            new_filename = filename.replace(' ', '')
            # Create full paths for the old and new filenames
            old_filepath = os.path.join(directory, filename)
            new_filepath = os.path.join(directory, new_filename)
            # Rename the file
            os.rename(old_filepath, new_filepath)
            print(f"Renamed: {old_filepath} -> {new_filepath}")

# Directories containing the JPEG and XML files
jpeg_images_dir = './JPEGImages'
annotations_dir = './Annotations'

# Remove spaces from JPEG filenames
remove_spaces_from_filenames(jpeg_images_dir, '.jpg')

# Remove spaces from XML filenames
remove_spaces_from_filenames(annotations_dir, '.xml')

print("Finished renaming files.")
```
